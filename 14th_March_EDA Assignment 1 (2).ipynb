```python
9) is there any significant difference in ejection fraction between patients with and without daibetes??
```

```python
#patients who have diabetes and those who not have diabetes
non_diabetic = df[df['diabetes'] == 0]
diabetic = df[df['diabetes'] == 1]

#plot
plt.figure(figsize=(15,8))
sns.kdeplot(non_diabetic['ejection_fraction'], label = 'non_diabetic' , multiple = 'stack' )
sns.kdeplot(diabetic['ejection_fraction'], label = 'diabetic', multiple = 'stack' )
plt.xlabel('Ejection Fraction Level')
plt.ylabel('Density')
plt.title('ejection fraction between patients with and without diabetes')
plt.legend()
plt.show()
```

```python
# insights = 
Ejection fraction level is slghtly low in the patients without diabetes. 
```

```python

plt.figure(figsize=(8, 6))
plt.scatter(df['ejection_fraction'], df['diabetes'], alpha=0.5)
plt.xlabel('Ejection Fraction')
plt.ylabel('Diabetes (0: No, 1: Yes)')
plt.title('Scatter Plot of Ejection Fraction vs Diabetes Status')
plt.grid(True)
plt.show()

# Count patients with diabetes based on ejection fraction
diabetes_count = df[df['diabetes'] == 1].shape[0]
Without_diabetes_count = df[df['diabetes'] == 0].shape[0]
print(f"Number of patients with diabetes: {diabetes_count}")
print(f"Number of patients without diabetes: {Without_diabetes_count}")

```

```python
#patients who survied and those who not survived
non_survived= df[df['DEATH_EVENT'] == 0]
survived = df[df['DEATH_EVENT'] == 1]

#plot
plt.figure(figsize=(15,8))
sns.kdeplot(non_survived['serum_creatinine'], label = 'non_survived' , multiple = 'stack' )
sns.kdeplot(survived['serum_creatinine'], label = 'survived', multiple = 'stack' )
plt.xlabel('Serum_Creatinine Level')
plt.ylabel('Density')
plt.title(' serum creatinine level vary between patients who survived and those who did not')
plt.legend()
plt.show()
```
